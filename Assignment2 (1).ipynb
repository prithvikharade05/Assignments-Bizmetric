```python
square_list = [i * i for i in range(1, 11)]
print(square_list)
```

```python
even_list = [i for i in range(1, 51) if i % 2 == 0]
print(even_list)
```

```python
name_list = ["apple", "banana", "mango", "grapes"]
upper_list = [i.upper() for i in name_list]
print(upper_list)
```

```python
num_list = [-10, 5, 7, -3, 0, 12]
positive_list = [i for i in num_list if i > 0]
print(positive_list)
```

```python
tuple_list = [(i, i*i) for i in range(1, 6)]
print(tuple_list)
```

```python
text = "prithviraj kharade"
vowel_list = [i for i in text if i in "aeiou"]
print(vowel_list)
```

```python
num_list = [10, -5, 7, -2, 15, -9]
new_list = [0 if i < 0 else i for i in num_list]
print(new_list)
```

```python
word_list = ["apple", "banana", "cat", "mango"]
length_list = [len(i) for i in word_list]
print(length_list)
```

```python
words = ["Apple", "banana", "Avocado", "cat", "ant"]
filtered = [word for word in words if not word.lower().startswith('a')]
print(filtered)
```

```python
numbers = [1, 2, 3, 4, 5, 6]
result = ["even" if num % 2 == 0 else "odd" for num in numbers]
print(result)
```

```python
nums = [n for n in range(1, 101) if n % 3 == 0 and n % 5 == 0]
print(nums)
```

```python
table = [[i * j for j in range(1, 6)] for i in range(1, 6)]
for row in table:
    print(row)
```

```python
data = {"name": "John", "age": 25, "city": "Pune"}
keys_list = [key for key in data]
print(keys_list)
```

```python
text = "a1b2c3d4"
digits = [char for char in text if char.isdigit()]
print(digits)
```

```python
result = "".join([char for char in text if char != " "])
print(result)
```

```python
text = "programming"
duplicates = [char for char in set(text) if text.count(char) > 1]
print(duplicates)
```

```python
sentences = ["Python is fast", "Python is easy"]
words = [word for sentence in sentences for word in sentence.split()]
print(words)
```

```python
nums = [1, 2, 2, 3, 4, 4, 5]
unique = [n for i, n in enumerate(nums) if n not in nums[:i]]
print(unique)
```

```python
A = [1, 2, 3]
B = ['a', 'b']
pairs = [(x, y) for x in A for y in B]
print(pairs)
```
