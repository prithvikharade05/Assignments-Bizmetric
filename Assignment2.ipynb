```python
square_list = [i * i for i in range(1, 11)]
print(square_list)
```

```python
even_list = [i for i in range(1, 51) if i % 2 == 0]
print(even_list)
```

```python
name_list = ["apple", "banana", "mango", "grapes"]
upper_list = [i.upper() for i in name_list]
print(upper_list)
```

```python
num_list = [-10, 5, 7, -3, 0, 12]
positive_list = [i for i in num_list if i > 0]
print(positive_list)
```

```python
tuple_list = [(i, i*i) for i in range(1, 6)]
print(tuple_list)
```

```python
text = "prithviraj kharade"
vowel_list = [i for i in text if i in "aeiou"]
print(vowel_list)
```

```python
num_list = [10, -5, 7, -2, 15, -9]
new_list = [0 if i < 0 else i for i in num_list]
print(new_list)
```

```python
word_list = ["apple", "banana", "cat", "mango"]
length_list = [len(i) for i in word_list]
print(length_list)
```
