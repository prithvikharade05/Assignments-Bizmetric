```python
# Question 12

salary = float(input("Enter yearly salary (in LPA): "))
rating = input("Enter rating (A/B/C/D): ").upper()

increment = 0

# Salary upto 5 LPA
if salary <= 5:
    if rating == 'A':
        increment = 16
    elif rating == 'B':
        increment = 12
    elif rating == 'C':
        increment = 10
    elif rating == 'D':
        increment = 6

# Salary upto 10 LPA
elif salary <= 10:
    if rating == 'A':
        increment = 14
    elif rating == 'B':
        increment = 10
    elif rating == 'C':
        increment = 8
    elif rating == 'D':
        increment = 6

# Salary upto 15 LPA
elif salary <= 15:
    if rating == 'A':
        increment = 8
    elif rating == 'B':
        increment = 6
    elif rating == 'C':
        increment = 4
    elif rating == 'D':
        increment = 0

# Salary upto 23 LPA
elif salary <= 23:
    if rating == 'A':
        increment = 7
    elif rating == 'B':
        increment = 5
    elif rating == 'C':
        increment = 4
    elif rating == 'D':
        increment = 0

else:
    print("Salary above defined range")

# Calculate new salary
if increment > 0:
    new_salary = salary + (salary * increment / 100)
    print("Increment =", increment, "%")
    print("New Salary =", new_salary, "LPA")

elif increment == 0 and salary <= 23:
    print("No increment given")
```

```python
def get_book_price(std, book):

    book_prices = {
        "Hindi":  {"1-4": 60, "5-8": 100, "9-10": 150},
        "Marathi":{"1-4": 60, "5-8": 100, "9-10": 150},
        "English":{"1-4": 80, "5-8": 100, "9-10": 150},
        "Science":{"1-4": 90, "5-8": 120, "9-10": 200},
        "Maths":  {"1-4": 100,"5-8": 140, "9-10": 250}
    }

    return book_prices[book][std]

def get_notebook_price(nb_type, pages):

    notebook_prices = {
        "square": {"100": 40, "200": 70},
        "4lines": {"100": 30, "200": 50},
        "2lines": {"100": 30, "200": 50},
        "single lines": {"100": 60, "200": 100},
        "A4": {"100": 100, "200": 180}
    }

    return notebook_prices[nb_type][pages]

std = input("Enter Standard (1-4 / 5-8 / 9-10): ")
book = input("Enter Book Name: ")
book_qty = int(input("Enter Book Quantity: "))
nb_type = input("Enter Notebook Type: ")
pages = input("Enter Pages (100 / 200): ")
nb_qty = int(input("Enter Notebook Quantity: "))
book_price = get_book_price(std, book) * book_qty
nb_price = get_notebook_price(nb_type, pages) * nb_qty

total = book_price + nb_price

print("\n----- BILL -----")
print("Book Cost =", book_price)
print("Notebook Cost =", nb_price)
print("Total Cost =", total)
```

```python
from datetime import date

def calculate_age(year, month, day):
    today = date.today()
    age = today.year - year

    if (today.month, today.day) < (month, day):
        age = age - 1

    return age

print(calculate_age(2001, 6, 5))
```

```python
def common_elements(l1, l2):
    common = []
    for item in l1:
        if item in l2 and item not in common:
            common.append(item)
    return common
l1 = [1, 2, 3, 4, 5]
l2 = [3, 2, 8, 7, 9]
print(common_elements(l1, l2))
```
